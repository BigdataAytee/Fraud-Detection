# **Detecting Fraudulent Online transaction**

The advent of online payment systems has enhanced transaction convenience and speed but also led to an increase in fraud, especially affecting credit card users. Detecting fraudulent transactions is crucial for credit card companies to protect customers from unauthorized charges. Using machine learning and Python, we can build a model to distinguish between legitimate and fraudulent transactions. We'll use a detailed dataset with transaction information and fraud labels to identify patterns indicative of fraud. This guide will walk you through data exploration, preprocessing, feature engineering, and model development to create an effective fraud detection system.

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns

# reading out the data for analysis
file_ = pd.read_csv('fraud.csv')
df= file_.copy()
df.head()
```

```python
#cheking the data information
df.info()
```

```python
# checking te summary statistics
df.describe()
```

```python
#checking for null data

df.isnull().sum()
```

```python
# dropping file with eempty sets
df.dropna(inplace= True)
```

```python
# encoding the type
from sklearn.preprocessing import LabelEncoder, StandardScaler
le= LabelEncoder()
df['type']= le.fit_transform(df['type'])
```

```python
# ploting to see if our target variable is balanced

plt.figure(figsize=(10,8))
plt.hist(df['isFraud'].unique())
plt.title('isFraud Distribution')
plt.show()
```

This above histogram indicates that our target variable is balanced. Therefore, we can proceed with our analysis

```python
# selecting the data for the model trainin
x= df[['type', 'amount', 'oldbalanceOrg', 'newbalanceOrig']]
y = df[['isFraud']]

# scaliing the x data
scaler = StandardScaler()
scaled_x = scaler.fit_transform(x)
# splitin to train and tesst the data
x_train, x_test, y_train, y_test = train_test_split(scaled_x,y, test_size= 0.2, random_state =42)

# training the model
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
# predicting with the model
y_pred= model.predict(x_test)
```

```python
# evaluating the model
from sklearn.metrics import accuracy_score
score = accuracy_score(y_pred, y_test)
score
```

```python
# testing to see if the classifierr works
#transaction_details= [type, amount, oldbalanceOrg, newbalanceOrig]
transaction_details = np.array([[3, 10000.51, 15000.20, 0.0]])

result =model.predict(transaction_details)
if result ==1:
  print('This is a fraudulent transaction')
else:
  print('This is not a Fraudulent transaction')
```
